## Conectando a conta do MQL5

```python
def autenticar_mql():
    login, password, server = open("secret").read().split()

    mt5.initialize(login=int(login), password=password, server=server)

    # estabelecemos a conexão com o terminal MetaTrader 5 para a conta especificada
    if not mt5.initialize(login=int(login), password=password, server=server):
        print("initialize() failed, error code =",mt5.last_error())
        quit()
    
    # imprimimos informações sobre o estado da conexão, o nome do servidor e a conta de negociação
    print(mt5.terminal_info())
    # imprimimos informações sobre a versão do MetaTrader 5
    print(mt5.version())

autenticar_mql()
```

### Preparando os dados para treinamento

```python
df_petro4 = df_petro4.dropna()
df_petro4 = df_petro4.drop_duplicates()
```

```python
# Definando o tamanho do treinamento
tamanho_dados_treinamento = int(len(df_petro4) * 0.8)

print(tamanho_dados_treinamento)

tamanho_dados_teste = int(len(df_petro4) - tamanho_dados_treinamento)

print(tamanho_dados_teste)
```

```python
df_petro4['recommendateion'] = 'Hold'

buy_condition = (df_petro4['MACD'] < df_petro4['MACD_signal']) & (df_petro4['RSI'] < 30)
sell_condition = (df_petro4['MACD'] > df_petro4['MACD_signal']) & (df_petro4['RSI'] > 70)

df_petro4.loc[buy_condition, 'recommendateion'] = 'Buy'
df_petro4.loc[sell_condition, 'recommendateion'] = 'Sell'

# Imprimir o DataFrame com as recomendações
print(df_petro4[['close', 'RSI', 'MACD', 'MACD_signal', 'recommendateion']])
```
